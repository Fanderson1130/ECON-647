```julia
using Plots

#variable values
R = 0.04
j = 1
D = 1
T = 100

#Prices functions
price_a(t)=D*(1-exp(-R*(T-t)))/R
g(x) = R*x - D

#Output arrays
p_ODE = Array{Float64}(101)
p_actual = Array{Float64}(101)
p_f = Array{Float64}(1)

periods = [0:j:99;]
p_f
p_actual = price_a(periods)

#starting values
p_f = 0

#Now we employ the Rugne_Kutta second order method, where alpha = .5
for t in periods
    p_f = p_f - j*g(p_f-j/2*g(p_f))
    p_ODE[T-t] = p_f
end

plot(p_actual, title = "PS3 Problem 3 (B)", label = "Actual Price")
plot!(p_ODE, label = "Estimated Price")

```
